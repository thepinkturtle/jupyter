```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
from jupyterthemes import get_themes
import jupyterthemes as jt
from jupyterthemes.stylefx import set_nb_theme
sns.set_style('darkgrid')

data = pd.read_csv(os.getcwd() + '/who_suicide_statistics.csv')
df = pd.DataFrame(data)
print(df.country.unique())

def plotSuicides(countrySelection):
    vis1 = sns.lmplot(data=df[df.country == countrySelection], x='year', y='suicides_no', hue='age',
              fit_reg=False, legend=False, height=8, aspect=2, scatter_kws={'s':130})
    vis1.set(xlabel='Year', ylabel='No. of Suicides', 
             title='Suicide trend for ' + countrySelection)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.show()

plotSuicides('United States of America')
```
